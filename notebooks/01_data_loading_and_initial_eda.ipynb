# 01. Загрузка данных и первичный EDA

Цель ноутбука — загрузить исходные таблицы датасета Olist, проверить их структуру, размеры, типы данных, пропуски, дубликаты и подготовить основу для дальнейшего анализа.

## План ноутбука

1. Импорт библиотек.
2. Загрузка CSV-файлов из `data/raw/`.
3. Проверка размеров таблиц.
4. Проверка типов данных.
5. Анализ пропусков.
6. Проверка дубликатов.
7. Первичные выводы по структуре данных.
